```python
class TreeNode:
    def __init__(self, name):  # Inisialisasi node
        self.name = name  # inisialisasi Nama dari node
        self.children = []  # List untuk menampung anak-anak node
        self.parent = None  # Parent node, diinisialisasi dengan None

    def add_child(self, child):  # Fungsi untuk menambah anak ke node
        child.parent = self  # Menjadikan node saat ini sebagai parent dari child
        self.children.append(child)  # Menambah child ke dalam daftar children

    def print_tree(self):  # Fungsi untuk mencetak TreeNode
        level = self.get_level()  # Mendapatkan level node saat ini
        space = ''  # Inisialisasi spasi untuk penampilan tree
        if level > 0:  # Jika level lebih dari 0, tambahkan spasi dan garis untuk menampilkan struktur tree
            space = " " * level + "|" + "_" * level
        print(space + self.name)  # Cetak nama node dengan spasi yang sesuai level
        if len(self.children) > 0:  # Jika ada anak-anak, cetak mereka juga
            for child in self.children:  # Rekursif panggil print_tree untuk setiap anak
                child.print_tree()

    def get_level(self):  # Fungsi untuk mendapatkan level node saat ini
        level = 0  # Inisialisasi level dengan 0
        p = self.parent  # Mulai dengan parent node
        while p:  # Iterasi ke atas sampai root
            level += 1  # Tambah level
            p = p.parent  # Pindah ke parent berikutnya
        return level  # Kembalikan level

    def find_path(self, destination, path=None):  # Fungsi untuk menemukan jalur ke tujuan
        if path is None:  # Jika path belum ada, inisialisasi dengan list kosong
            path = []

        path.append(self.name)  # Tambahkan nama node saat ini ke dalam path

        if self.name == destination:  # Jika node saat ini adalah tujuan, kembalikan path
            return path

        for child in self.children:  # Lakukan pencarian rekursif pada setiap anak
            result = child.find_path(destination, path.copy())  # Gunakan path.copy() untuk membuat salinan path yang berbeda
            if result:  # Jika ditemukan jalur yang benar, kembalikan hasilnya
                return result

        return None  # Jika tidak ditemukan, kembalikan None


    def build_tree():  # Fungsi statis untuk membangun tree
        # Membuat node root dan node lainnya
        surabaya = TreeNode('Surabaya Pasarturi')  # root
        gubeng = TreeNode('Gubeng')
        wonokromo = TreeNode('Wonokromo')
        sepanjang = TreeNode('Sepanjang')
        krian = TreeNode('Krian')
        waru = TreeNode('Waru')
        gedang = TreeNode('Gedangan')
        sidoarjo = TreeNode('Sidoarjo')
        tanggulangin = TreeNode('Tanggulangin')
        porong = TreeNode('Porong')
        bangil = TreeNode('Bangil')
        tarik = TreeNode('Tarik')
        mojokerto = TreeNode('Mojokerto')

        # Menyusun tree dengan mengatur parent-child relationships
        surabaya.add_child(gubeng)
        gubeng.add_child(wonokromo)
        wonokromo.add_child(sepanjang)
        wonokromo.add_child(waru)
        sepanjang.add_child(krian)
        krian.add_child(tarik)
        tarik.add_child(mojokerto)
        waru.add_child(gedang)
        gedang.add_child(sidoarjo)
        sidoarjo.add_child(tanggulangin)
        tanggulangin.add_child(porong)
        porong.add_child(bangil)

        return surabaya  # Kembalikan root tree

def find_routes(start, destinations):  # Fungsi untuk menemukan rute ke beberapa tujuan
    tree = TreeNode.build_tree()  # Membangun tree
    results = {}  # Inisialisasi dictionary untuk menyimpan hasil
    for destination in destinations:  # Iterasi melalui setiap tujuan
        path = tree.find_path(destination)  # Mencari jalur ke tujuan
        if path:  # Jika ditemukan jalur
            results[destination] = ' -> '.join(path)  # Simpan jalur yang ditemukan dalam format string
        else:  # Jika tidak ditemukan
            results[destination] = f'Node {destination} tidak ditemukan dalam tree'  # Simpan pesan error
    return results  # Kembalikan hasil pencarian

if __name__ == '__main__':  # Jika skrip dijalankan secara langsung
    root = TreeNode.build_tree()  # Membangun tree dan menyimpan root
    root.print_tree()  # Mencetak struktur tree

    # Mendapatkan input dari pengguna
    user_input = input("Masukkan destinasi yang diinginkan (jika lebih dari satu destinasi pisahkan dengan koma): ")
    destinations = [dest.strip() for dest in user_input.split(',')]  # Memisahkan input berdasarkan koma dan menghapus spasi tambahan

    # Menemukan dan mencetak rute untuk tujuan yang diberikan
    routes = find_routes('Surabaya Pasarturi', destinations)  # Menemukan rute dari Surabaya Pasarturi ke tujuan yang diberikan
    for dest, path in routes.items():  # Iterasi melalui hasil pencarian rute
        print(f'Route to {dest}: {path}')  # Mencetak setiap rute yang ditemukan
```
